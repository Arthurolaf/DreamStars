 # Непрерывный генетический алгоритм с нуля с Python #

 Генетический алгоритм - это мощная техника оптимизации, вдохновленная природой. Генетические алгоритмы имитируют эволюцию, чтобы найти лучшее решение. В отличие от большинства алгоритмов оптимизации, генетические алгоритмы не используют производные для поиска минимумов. Одним из наиболее значительных преимуществ генетических алгоритмов является их способность находить глобальный минимум, не застревая в локальных минимумах. Случайность играет существенную роль в структуре генетических алгоритмов, и именно поэтому генетические алгоритмы продолжают искать в пространстве поиска. Непрерывность в заголовке означает, что генетический алгоритм, который мы собираемся создать, будет использовать плавающие числа или целые числа в качестве параметров оптимизации вместо двоичных чисел.

![''](/Arthurolaf/DreamStars/blob/main/AI_algaritms/img/Block-schema.png)


Генетические алгоритмы создают начальную популяцию случайно сгенерированных решений-кандидатов, эти варианты-кандидаты оцениваются и вычисляется их пригодность. 

Значение пригодности решения - это числовое значение, которое определяет, насколько хорошо решение, чем выше значение пригодности, тем лучше решение.

Если исходная популяция не соответствует требованиям критериев прекращения, генетический алгоритм создает следующее поколение.
* Первая генетическая операция - Селекция; 
в этой операции выбираются люди, которые собираются перейти к следующему поколению.

* После процесса выбора начинается операция сопряжения.
* Операция спаривания соединяет выбранных особей два на два для операции Спаривания. Операция «Сопряжение» берет парных родительских особей и создает потомков, которые будут заменять особей, которые не были выбраны в операции «Выбор», поэтому в следующем поколении будет то же число особей, что и в предыдущем поколении. Этот процесс повторяется до тех пор, пока не будут выполнены критерии прекращения.

```python
def individual(number_of_genes, upper_limit, lower_limit):
    individual=[round(rnd()*(upper_limit-lower_limit)
                +lower_limit,1) for x in range(number_of_genes)]
    return individual
```

<h4> выбор </h4>

Функция выбора берет совокупность возможных решений и их значения пригодности (поколение) и выводит людей, которые собираются перейти к следующему поколению.<br>
Элитарность может быть введена в генетический алгоритм, который автоматически выберет лучшего человека в поколении, поэтому мы не теряем лучшее решение.<br>
Есть несколько методов выбора, которые можно использовать. Методы выбора, приведенные в этой статье:
* **Выбор колеса рулетки:**
В выборе колеса рулетки у каждого человека есть шанс быть выбранным. Вероятность выбора индивидуума зависит от его пригодности. У более подходящих людей есть более высокий шанс быть выбранным.
![''](/Arthurolaf/DreamStars/blob/main/AI_algaritms/img/exemple_table2.png)

Функция выбора колеса рулетки принимает кумулятивные суммы и случайно сгенерированное значение для процесса выбора и возвращает номер выбранного человека. Посредством вычисления кумулятивных сумм каждый индивид имеет уникальное значение от 0 до 1. Чтобы выбрать индивидов, число от 0 до 1 генерируется случайным образом и выбирается индивид, близкий к случайно сгенерированному числу. Функция рулетки может быть записана как;


```python
def roulette(cum_sum, chance):
    veriable = list(cum_sum.copy())
    veriable.append(chance)
    veriable = sorted(veriable)
    return veriable.index(chance)
```

* **Выбор наиболее подходящей половины:** В этом методе выбора выбирается наиболее подходящая половина возможных решений для перехода к следующему поколению.
![''](../AI_algaritms/img/exemple_table3.png)

* **Случайный выбор** :В этом методе индивидуумы отбираются случайным образом.

![''](../AI_algaritms/img/exemple_table4.png)

Функция выбора может быть записана как;



```python
def selection(generation, method='Fittest Half'):
    generation['Normalized Fitness'] = \
        sorted([generation['Fitness'][x]/sum(generation['Fitness']) 
        for x in range(len(generation['Fitness']))], reverse = True)
    generation['Cumulative Sum'] = np.array(
        generation['Normalized Fitness']).cumsum()
    
    if method == 'Roulette Wheel':
        selected = []
        for x in range(len(generation['Individuals'])//2):
            selected.append(roulette(generation
                ['Cumulative Sum'], rnd()))
            while len(set(selected)) != len(selected):
                selected[x] = \
                    (roulette(generation['Cumulative Sum'], rnd()))
        selected = {'Individuals': 
            [generation['Individuals'][int(selected[x])]
                for x in range(len(generation['Individuals'])//2)]
                ,'Fitness': [generation['Fitness'][int(selected[x])]
                for x in range(
                    len(generation['Individuals'])//2)]}
    
    elif method == 'Fittest Half':
        selected_individuals = [generation['Individuals'][-x-1]
            for x in range(int(len(generation['Individuals'])//2))]
        selected_fitnesses = [generation['Fitness'][-x-1]
            for x in range(int(len(generation['Individuals'])//2))]
        selected = {'Individuals': selected_individuals,
                    'Fitness': selected_fitnesses}
    
    elif method == 'Random':
        selected_individuals = \
            [generation['Individuals']
                [randint(1,len(generation['Fitness']))]
            for x in range(int(len(generation['Individuals'])//2))]
        selected_fitnesses = [generation['Fitness'][-x-1]
            for x in range(int(len(generation['Individuals'])//2))]
        selected = {'Individuals': selected_individuals,
                    'Fitness': selected_fitnesses}
    return selected
```

<h4> Сопряжение </h4>

Спаривание и сопряжение используются как одна операция в большинстве приложений генетического алгоритма, но для создания более простых функций и возможности простого использования различных алгоритмов спаривания и сопряжения в этом приложении две генетические операции разделены. Если в генетическом алгоритме есть элитарность, элита должна быть входом для функции, а также для выбранных людей. Мы собираемся обсудить три различных метода сопряжения;

* **Fittest:** В этом методе люди объединяются два в два, начиная с самого подходящего человека. Таким образом, более подходящие люди соединяются вместе, но менее подходящие люди также соединяются вместе.

![''](../AI_algaritms/img/exemple_table5.png)

Лучшая фигура
* **Случайная:**В этом методе люди попарно соединяются два на два.

![''](../AI_algaritms/img/exemple_table6.png)

* **Взвешенная случайная величина:** В этом методе индивиды спариваются случайным образом два на два, но более подходящие индивидуумы имеют более высокий шанс быть выбранными для спаривания.

![''](../AI_algaritms/img/exemple_table7.png)


Функция сопряжения может быть записана как;




```python
def pairing(elit, selected, method = 'Fittest'):
    individuals = [elit['Individuals']]+selected['Individuals']
    fitness = [elit['Fitness']]+selected['Fitness']
    
    if method == 'Fittest':
        parents = [[individuals[x],individuals[x+1]] 
                   for x in range(len(individuals)//2)]
    
    if method == 'Random':
        parents = []
        for x in range(len(individuals)//2):
            parents.append(
                [individuals[randint(0,(len(individuals)-1))],
                 individuals[randint(0,(len(individuals)-1))]])
            while parents[x][0] == parents[x][1]:
                parents[x][1] = individuals[
                    randint(0,(len(individuals)-1))]
    
    if method == 'Weighted Random':
        normalized_fitness = sorted(
            [fitness[x] /sum(fitness) 
             for x in range(len(individuals)//2)], reverse = True)
        cummulitive_sum = np.array(normalized_fitness).cumsum()
        parents = []
        for x in range(len(individuals)//2):
            parents.append(
                [individuals[roulette(cummulitive_sum,rnd())],
                 individuals[roulette(cummulitive_sum,rnd())]])
            while parents[x][0] == parents[x][1]:
                parents[x][1] = individuals[
                    roulette(cummulitive_sum,rnd())]
    return parents
```

<h4> Спаривание </h4>

Мы обсудим два разных метода спаривания. В приведенном ниже коде Python два отобранных родительских особа создают двух потомков. Есть два метода спаривания, которые мы собираемся обсудить.

* **Отдельная точка:** В этом методе гены после одной точки заменяются генами другого родителя, чтобы создать двух потомков.

![''](../AI_algaritms/img/exemple_table8.png)

* **Два момента:** В этом методе гены между двумя точками заменяются генами другого родителя, чтобы создать двух потомков.

![''](../AI_algaritms/img/exemple_table9.png)



```python
def mating(parents, method='Single Point'):

    if method == 'Single Point':
        pivot_point = randint(1, len(parents[0]))
        offsprings = [parents[0] \
            [0:pivot_point]+parents[1][pivot_point:]]
        offsprings.append(parents[1]
            [0:pivot_point]+parents[0][pivot_point:])
        
    if method == 'Two Pionts':
        pivot_point_1 = randint(1, len(parents[0]-1))
        pivot_point_2 = randint(1, len(parents[0]))
        while pivot_point_2<pivot_point_1:
            pivot_point_2 = randint(1, len(parents[0]))
        offsprings = [parents[0][0:pivot_point_1]+
            parents[1][pivot_point_1:pivot_point_2]+
            [parents[0][pivot_point_2:]]]
        offsprings.append([parents[1][0:pivot_point_1]+
            parents[0][pivot_point_1:pivot_point_2]+
            [parents[1][pivot_point_2:]]])

    return offsprings
```

<h4> Мутации </h4>

Последняя генетическая операция - случайные мутации. Случайные мутации происходят у отобранных особей и их потомков, чтобы улучшить разнообразие следующего поколения. Если в генетическом алгоритме есть элитарность, элитный индивид не проходит случайных мутаций, поэтому мы не теряем лучшего решения. Мы собираемся обсудить два разных метода мутации.

* **Гаусс:** В этом методе ген, который проходит мутацию, заменяется числом, которое генерируется в соответствии с распределением Гаусса вокруг исходного гена.
* **Сброс:** В этом методе исходный ген заменяется случайно сгенерированным геном.




```python
def mutation(individual, upper_limit, lower_limit, muatation_rate=2, 
    method='Reset', standard_deviation = 0.001):
    gene = [randint(0, 7)]
    for x in range(muatation_rate-1):
        gene.append(randint(0, 7))
        while len(set(gene)) < len(gene):
            gene[x] = randint(0, 7)
    mutated_individual = individual.copy()
    if method == 'Gauss':
        for x in range(muatation_rate):
            mutated_individual[x] = \
            round(individual[x]+gauss(0, standard_deviation), 1)
    if method == 'Reset':
        for x in range(muatation_rate):
            mutated_individual[x] = round(rnd()* \
                (upper_limit-lower_limit)+lower_limit,1)
    return mutated_individual
```

# Создание следующего поколения #

Следующее поколение создано с использованием генетических операций, которые мы обсуждали. 

Элитарность может быть введена в генетический алгоритм при создании следующего поколения. 

Элитарность - это код Python для создания следующего поколения:



```python
def next_generation(gen, upper_limit, lower_limit):
    elit = {}
    next_gen = {}
    # берем лучших из  списка индивидумов
    elit['Individuals'] = gen['Individuals'].pop(-1)
    # Берем лучшие показатели фитнеса
    elit['Fitness'] = gen['Fitness'].pop(-1)
    
    # Приизводим селекцию более подходящих 
    selected = selection(gen)
    
    # Образуем пары из получившегося списка лучших и подходящих
    parents = pairing(elit, selected)
    
    # Скриещиваем и получаем потомков.
    offsprings = [[[mating(parents[x])
                    for x in range(len(parents))]
                    [y][z] for z in range(2)] 
                    for y in range(len(parents))]
    offsprings1 = [offsprings[x][0]
                   for x in range(len(parents))]
    offsprings2 = [offsprings[x][1]
                   for x in range(len(parents))]
    
    # Собираем оставшихся кроме элиты и две новых особи в общий набор
    unmutated = selected['Individuals']+offsprings1+offsprings2
    
    # Производим мутацию в генах
    mutated = [mutation(unmutated[x], upper_limit, lower_limit) 
        for x in range(len(gen['Individuals']))]
    
    # добавляем элитку
    unsorted_individuals = mutated + [elit['Individuals']]
    
    # Снова всех выстраиваем 1-ый, 2-ой по фитнесу и сортеруем их включая Элиту 
    unsorted_next_gen = \
        [fitness_calculation(mutated[x]) 
         for x in range(len(mutated))]
    unsorted_fitness = [unsorted_next_gen[x]
        for x in range(len(gen['Fitness']))] + [elit['Fitness']]
    sorted_next_gen = \
        sorted([[unsorted_individuals[x], unsorted_fitness[x]]
            for x in range(len(unsorted_individuals))], 
                key=lambda x: x[1])
    next_gen['Individuals'] = [sorted_next_gen[x][0]
        for x in range(len(sorted_next_gen))]
    next_gen['Fitness'] = [sorted_next_gen[x][1]
        for x in range(len(sorted_next_gen))]
    gen['Individuals'].append(elit['Individuals'])
    gen['Fitness'].append(elit['Fitness'])
    return next_gen
```

# Критерии прекращения #

После создания поколения критерии завершения используются для определения того, должен ли генетический алгоритм создать другое поколение или он должен остановиться. Различные критерии завершения могут использоваться одновременно, и если генетический алгоритм удовлетворяет одному из критериев, генетический алгоритм останавливается. Мы собираемся обсудить четыре критерия прекращения

* **Максимальный фитнес:** Этот критерий прекращения проверяет, удовлетворяет ли наш самый подходящий человек в текущем поколении нашим критериям. Используя этот метод термиантирования, можно получить желаемые результаты. Как видно из рисунка ниже, максимальный предел пригодности может быть определен с учетом некоторых локальных минимумов.
* **Максимальный средний фитнес:** Если нас интересует набор решений, можно проверить средние значения людей в нынешних поколениях, чтобы определить, удовлетворяет ли нынешнее поколение нашим ожиданиям.
* **Максимальное количество поколений:** Мы могли бы ограничить максимальное количество поколений, созданных генетическим алгоритмом.
* **Максимальный аналогичный номер фитнеса:** Благодаря элитарности лучший человек в поколении переходит к следующему поколению, не мутируя. Этот человек может быть лучшим человеком в следующем поколении. Мы можем ограничить число для того же человека, чтобы быть лучшим человеком, поскольку это можно петь, что генетический алгоритм застрял в локальных минимумах. Функция проверки, изменились ли максимальные значения пригодности, может быть записана как;



```python
def fitness_similarity_chech(max_fitness, number_of_similarity):
    print(max_fitness, number_of_similarity)
    result = False
    similarity = 0
    for n in range(len(max_fitness)-1):
        if max_fitness[n] == max_fitness[n+1]:
            similarity += 1
        else:
            similarity = 0
    if similarity == number_of_similarity-1:
        result = True
    return result
```
